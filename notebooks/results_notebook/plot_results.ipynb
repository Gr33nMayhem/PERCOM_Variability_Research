```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
run_log_directory = '../../data/Run_logs/'
sliding_window_directory = '../../data/Sliding_window/'
devices = ['empatica-left',
           'empatica-right',
           'bluesense-LUA',
           'bluesense-LWR',
           'bluesense-RUA',
           'bluesense-RWR1',
           'bluesense-RWR2',
           'bluesense-TRS',
           'maxim-red',
           'maxim-green', ]
for device in devices:
    parent_directory = run_log_directory + device + '/logs/'
    if not os.path.exists(parent_directory):
        continue
    # get all the sub directories in the parent directory
    sub_directories = os.listdir(parent_directory)
    for directory in sub_directories:
        cv_directories = os.listdir(parent_directory + directory + '/')
        for cv_directory in cv_directories:
            # check if a directory
            if not os.path.isdir(parent_directory + directory + '/' + cv_directory):
                continue
            for other_device in devices:
                print(device)
                print(other_device)
                # open the csv file prediction_result_ and read the data
                csv_file = parent_directory + directory + '/' + cv_directory + '/prediction_result_' + device + '_' + other_device + '.csv'
                # open with pandas
                df = pd.read_csv(csv_file)
                # if the file is empty, continue
                if df.empty:
                    continue

                # get the prediction result
                preds = df['preds'].values
                # get the ground truth
                trues = df['trues'].values

                # plot a confusion matrix
                cm = confusion_matrix(trues, preds)
                # plot the confusion matrix using seaborn
                sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
                plt.xlabel('Predicted')
                plt.ylabel('True')
                plt.title('Confusion Matrix for ' + device)
                # check if folder graph exists, if not create it
                if not os.path.exists(parent_directory + directory + '/' + cv_directory + '/graphs'):
                    os.makedirs(parent_directory + directory + '/' + cv_directory + '/graphs')
                plt.savefig(
                    parent_directory + directory + '/' + cv_directory + '/graphs' + '/confusion_matrix_' + device + '_' + other_device + '.png')
                plt.close()
                print(directory)
                print(cv_directory)

                # print the accuracy
                print('Accuracy for ' + device + ' is ' + str(np.sum(preds == trues) / len(preds)))
                # print the precision
                print(
                    'Precision for ' + device + ' is ' + str(np.sum((preds == 1) & (trues == 1)) / np.sum(preds == 1)))
                # print the recall
                print('Recall for ' + device + ' is ' + str(np.sum((preds == 1) & (trues == 1)) / np.sum(trues == 1)))
                # print the f1 score
                print('F1 score for ' + device + ' is ' + str(
                    2 * np.sum((preds == 1) & (trues == 1)) / (np.sum(preds == 1) + np.sum(trues == 1))))
                print('-----------------------------------------')
```
